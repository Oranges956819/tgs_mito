```python
import numpy as np

import matplotlib.pyplot as plt
import matplotlib as mplot

matrix = np.loadtxt("filelog_2024-08-21_10:48.txt", dtype=str, delimiter='\t', skiprows=1, unpack='true')

n_cols = len(matrix[:,0])
n_rows = len(matrix[0,:])
print('Number of rows:' + str(n_rows), end='\t')
print('Number of cols:' + str(n_cols), end='\n\n')
elevation = np.zeros(n_rows,float)
azimuth = np.zeros(n_rows,float)

for i in range(n_rows):
    elevation[i] = float(matrix[1,i])
    azimuth[i] = float(matrix[2,i])


for i in range(n_rows-60):
    
    if(i%2==0):
        plt.plot(azimuth[i],elevation[i],'r.')
        if(azimuth[i]>161.5):
            plt.text(azimuth[i]+0.03,elevation[i],str(i+1))
        else:
            plt.text(azimuth[i]-0.15,elevation[i],str(i+1))
        plt.plot([azimuth[i],azimuth[i+1]],[elevation[i],elevation[i+1]], color='b', linestyle='--')
    else:
        plt.plot(azimuth[i],elevation[i],'b.')
        if(azimuth[i]>161.5):
            plt.text(azimuth[i]+0.03,elevation[i],str(i+1))
        else:
            plt.text(azimuth[i]-0.15,elevation[i],str(i+1))
        plt.plot([azimuth[i],azimuth[i+1]],[elevation[i],elevation[i+1]], color='r', linestyle='--')
        
    #if(i%2==0):
     #   plt.plot(azimuth[i],elevation[i],'r.')
      #  plt.plot(, color='r', linestyle='-.')
    #else:
     #   plt.plot(azimuth[i],elevation[i],'b.')
        #plt.axhline(y=dec, color='b', linestyle='-.')
    


plt.grid()
plt.show()
```
